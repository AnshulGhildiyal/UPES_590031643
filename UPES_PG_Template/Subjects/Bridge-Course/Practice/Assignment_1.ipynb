### *Level Up -- Peeking inside the box*

The textbook question stops at "print the value and its type." I wanted to know: *how does Python actually store these differently in memory, and does `a = b` copy the box or just make a second label for the same box?* This distinction (identity vs equality, mutable vs immutable) trips up almost every beginner sooner or later, so it's worth seeing directly rather than being told about it.

````python
import sys

def inspect_value(label, value):
    print(f"{label:10s} | type ={type(value).__name__:6s} | id={id(value)} | size={sys.getsizeof(value)}B")

a = 10
b = 10           # small integers are cached in CPython, so a and b likely points to the same object
c = 10.0
d = [1, 2, 3]
e = d            # e is just another reference to the same list as d
f = list(d)      # f is a new list containing the same elements as d

inspect_value("a", a)
inspect_value("b", b)
print("a is b: ", a is b, "(CPython caches small ints -2 to 256 -- same box, two labels)")
inspect_value("c", c)
inspect_value("d", d)
inspect_value("e (=d)", e)
inspect_value("f (copy)", f)
print("d is e:", d is e, "  <- same object, mutating one mutates 'both'")
print("d is f:", d is f, "  <- different objects")
print("d == f:", d == f, " <- but equal VALUES")

d.append(99)
print("\nAfter d.append(99):")
print("d:", d, "| e:", e, " <- e changed too, because it's the SAME list")
print("f:", f, " <- f is untouched, it was a real copy")
````

**Why this matters in real code:** this exact gotcha (`e = d` vs `f = list(d)`) is one of the most common sources of bugs professional developers hit — passing a list into a function and being surprised the original changed outside it. Understanding it now, in Experiment 1, saves debugging pain in every experiment after this one.

Calculate Simple Interest for a loan, compute the sum of the first $n$ natural numbers, and convert raw seconds into standard hours/minutes/seconds format.

You just took out a student loan to buy a Macbook Air (Simple Interest). Now you are pulling an all-nighter, staring at a timer showing exactly 7,384 seconds until your 8 AM exam.
````python
# --- Q7: Student Loan (Simple Interest) ---
principal = 10000 
rate = 8          
time = 2          
simple_interest = (principal * rate * time) / 100 
print(f"Loan Interest Owed: ${simple_interest:.2f}") 

# --- Q11: Sum of first 'n' natural numbers ---
n = 10
# Using the mathematical formula instead of a slow loop: n(n+1)/2
sum_n = n * (n + 1) // 2
print(f"Sum of first {n} numbers: {sum_n}")

# --- Q9: The Exam Countdown (Time Conversion) ---
total_seconds = 7384

# Floor division (//) gets the solid hours. 
hours = total_seconds // 3600

# Modulo (%) gets what's left over AFTER extracting hours, then we divide by 60 for minutes
minutes = (total_seconds % 3600) // 60

# Modulo (%) gets the absolute final leftover seconds
seconds = total_seconds % 60

print(f"Time until exam: {hours} hour(s), {minutes} minute(s), {seconds} second(s)")
````

### ***Schedule Changes (Swapping & Membership)***
Swap the values of two variables without using a third temporary variable. Use the membership operator to check if an item exists within a sequence or a string

### The Analogy
- **Swapping:** Imagine you have a morning class (Variable A) and an evening class (Variable B). You want to trade them. Normally, you'd need a third "temporary" placeholder slot to hold one while you move the other. But using pure math (or Python magic), you can juggle them directly.
- **Membership:** You are checking a massive, crowded waitlist pinned to a door (a Tuple/Sequence) to see if your exact Student ID number is on it (`in`).

````python
# --- Q10: The Mathematical Swap (No temp variable) ---
a = 25 # Morning Class
b = 40 # Evening Class
print(f"Before math swap: a={a}, b={b}")

a = a + b # a is now 65
b = a - b # b is now 65 - 40 = 25 (Swapped!)
a = a - b # a is now 65 - 25 = 40 (Swapped!)
print(f"After math swap:  a={a}, b={b}") 

# --- The "Pythonic" Swap (Level Up) ---
# Under the hood, this uses Tuple Packing/Unpacking, bypassing math entirely.
a, b = b, a 
print(f"After Python swap: a={a}, b={b} (Back to normal)")

# --- Q14 & 15: The Waitlist Checker (Membership Operator) ---
waitlist_sequence = (10, 20, 56, 78, 89) 
student_number = 56                      
course_name = "Python Programming"       
search_char = "g"                        

# The 'in' keyword returns a Boolean (True/False)
is_enrolled = student_number in waitlist_sequence 
char_found = search_char in course_name          

print(f"Is student {student_number} on the waitlist? {is_enrolled}") 
print(f"Is '{search_char}' in '{course_name}'? {char_found}")        
````

- **Complexity:** Searching a tuple with `in` is $O(N)$ time complexity, as it has to scan each item one by one until it finds a match.

### ***The Late-Night CS Lab (Bitwise Operations)***
Generate a truth table for bitwise operators AND (&), OR (|), XOR (^), and demonstrate bitwise Left Shift and Right Shift

### The Analogy
You are sneaking into the campus server room. Bitwise operators don't do regular math. They are physical switches. They bypass the high-level numbers completely and violently flip the raw 1s and 0s (electricity on/off) inside the computer's memory.
- `A Left Shift (<<)` pushes all the electrical currents to the left, which is a lightning-fast way to multiply by 2.

````python
# --- Q12: Server Room Switches (Truth Table) ---
print("Truth Table for Bitwise Operators:")
print("a b | a&b (AND) | a|b (OR) | a^b (XOR)")
for a in (0, 1):
    for b in (0, 1):
        # & = Both must be 1.  | = Either is 1.  ^ = Must be DIFFERENT.
        print(f"{a} {b} |    {a & b}      |    {a | b}     |    {a ^ b}")

# --- Q13: Bit Shifting (Fast Math) ---
num = 10   # In binary: 1010
shift = 2

# Shifting left by 1 multiplies by 2. Shifting by 2 multiplies by 4 (10 * 4 = 40)
left_shift = num << shift 

# Shifting right by 1 floor-divides by 2. Shifting by 2 divides by 4 (10 // 4 = 2)
right_shift = num >> shift

print("\nBitwise Shifts:")
print(f"Original Num : {num:08b} (Base 10: {num})")
print(f"Left shift 2 : {left_shift:08b} (Base 10: {left_shift})")
print(f"Right shift 2: {right_shift:08b} (Base 10: {right_shift})")
````

### Under the hood:
- **Complexity:** $O(1)$. This is the absolute fastest operation a CPU can perform because there is no abstraction layer.
- **Gotchas:** Do not confuse bitwise `&` / `|` with logical `and` / `or`.
    - `True and False` evaluates Boolean logic.
    - `5 & 3` looks at the binary `0101 & 0011` and outputs `0001` (which is the integer 1).

Notice in the code I used the f-string formatting trick `{num:08b}`. This forces Python to display the variable as a raw 8-bit binary string so you can literally see the 1s and 0s shifting left and right in your terminal output.

### ***Edge-Case Tests***
What happens if you try to bit-shift a float? e.g., `3.14 << 1`. Python throws a `TypeError`. Bitwise operations operate strictly on memory integers.

## Question 5: Quadratic equation roots

**Task:** Check whether a quadratic equation has real or imaginary roots and display the roots.

### ***The Physics Lab Simulator***

### The Analogy
You are building a trajectory simulator for a physics class. The quadratic equation tracks the arc of a thrown object. The Discriminant ($b^2 - 4ac$) acts as the reality check. If it's positive, the object hit the ground in reality (Real Roots). If it's negative, the math breaks reality, meaning the object never crossed the ground axis (Imaginary Roots).

````python
import cmath 

# Equation format: ax^2 + bx + c = 0
a, b, c = 1, 2, 5

discriminant = b ** 2 - 4 * a * c

# cmath is used here because math.sqrt crashes on negative numbers
root1 = (-b + cmath.sqrt(discriminant)) / (2 * a)
root2 = (-b - cmath.sqrt(discriminant)) / (2 * a)

if discriminant >= 0:
    print("Roots are real")
else:
    print("Roots are imaginary")

print(f"Root 1: {root1}")
print(f"Root 2: {root2}")
````

### Under the hood
- **Complexity:** $O(1)$.
- **Gotchas:** Standard `math.sqrt(-4)` will throw a `ValueError` because you can't square root a negative number in basic arithmetic. Python provides the `cmath` (Complex Math) module specifically to handle imaginary numbers gracefully without crashing the pipeline.

### ***Level up***

### Why does Python check `% 400` for a leap year?

At first glance, this condition looks confusing:

```python
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
```

A common misconception is:

> "A leap year has **366 days**, so why are we dividing the **year** by **400**?"

The answer is that **Python is not checking the number of days in a year.**

It is checking **the year number itself** according to the rules of the **Gregorian Calendar**.

For example:

```python
year = 2024

print(year % 4)    # 0
print(year % 100)  # 24
print(year % 400)  # 24
```

The `%` (modulo) operator returns the **remainder** after division.

- `2024 % 4 == 0` → 2024 is divisible by 4 ✅
- `2024 % 100 != 0` → 2024 is **not** divisible by 100 ✅
- Therefore, **2024 is a leap year.**

---

### Why isn't every year divisible by 4 a leap year?

If the leap year rule were simply:

```python
if year % 4 == 0:
    print("Leap year")
```

then years like **1700, 1800, 1900, and 2100** would incorrectly become leap years.

The problem is that the Earth **doesn't orbit the Sun in exactly 365.25 days.**

It actually takes approximately:

```text
365.2422 days
```

Adding one extra day every four years introduces a small error that accumulates over centuries.

To correct this drift, the **Gregorian Calendar** adds two additional rules.

---

### The Gregorian Calendar Rules

***Rule 1***

Every year divisible by **4** is a leap year.

Examples:

- ✅ 2016
- ✅ 2020
- ✅ 2024

---

***Rule 2***

However, years divisible by **100** are **not** leap years.

Examples:

- ❌ 1700
- ❌ 1800
- ❌ 1900
- ❌ 2100

This removes the extra leap days that would slowly shift the calendar.

---

***Rule 3***

There's one final exception.

Years divisible by **400** **are** leap years.

Examples:

- ✅ 1600
- ✅ 2000
- ✅ 2400

This rule compensates for skipping too many leap years and keeps our calendar aligned with Earth's orbit.

````python
terms = 10
a, b = 0,1

print("Fibonacci series: ")
for _ in range(terms):
    print(a, end =" ")
    #Simultaneous variable swapping using tuple packing
    a, b = b, a + b
print()
````

### Under the hood:

- **Gotchas:** The underscore `_` in `for _ in range(terms)` is a Python convention. It tells other developers, "I need to loop exactly this many times, but I don't actually care about using the index variable itself."
- **Complexity:** $O(N)$ time complexity and $O(1)$ space complexity, as we only store the last two variables in memory regardless of how large `N` gets.

### ***Level up***
If you need to generate 100,000 terms for an AI dataset, returning a massive list will cause an out-of-memory error. DevOps architecture uses **Generators** (`yield`). Instead of computing the whole list at once, a generator yields one number, pauses its state in memory, and waits for you to ask for the next one.

````python
# The Generator Approach
def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# It only consumes memory for one item at a time
for number in fibonacci_generator(10):
    print(number, end=" ")
````

Because this experiment is fundamentally about **Loops**, how does `.upper()` actually work underneath the abstraction? It loops through the ASCII mathematical values of the characters. Lowercase 'a' is 97. Uppercase 'A' is 65. The difference is exactly 32.

````python
# The manual loop showing how the machine processes it
text = "Python Programming Lab"
manual_upper = ""

for char in text:
    # If the character is lowercase (ASCII 97 to 122)
    if 'a' <= char <= 'z':
        # Shift it mathematically by 32 down to the Uppercase ASCII block
        manual_upper += chr(ord(char) - 32)
    else:
        manual_upper += char

print(f"Manual ASCII Loop: {manual_upper}")
````

## Question 9: Prime numbers between 1 and 100

**Task:** Print all prime numbers between 1 and 100.

### ***The Study Room Search***

### The Analogy
It is finals week. You are walking through the library corridors (rooms 1 to 100) looking for completely empty study rooms. If even one person (a divisor) is in the room, it’s compromised, and you move on to the next. You need a nested system: walking down the hall (Outer Loop), and looking inside each room (Inner Loop).

````python
primes = []

for num49 in range(2, 101):
    is_prime49 = True

    for i in range(2, int(num49 ** 0.5) + 1):
        if num49 % i == 0:
            is_prime49 = False
            break

    if is_prime49:
        primes.append(num49)

print(f"Prime numbers from 1 to 100 are : {primes}")
````

### Under the hood:

- **Gotchas:** The `is_prime` boolean flag must be reset to `True` at the start of *every* iteration of the outer loop. If you place it outside the loops, the moment one number fails, all subsequent numbers will be falsely marked as non-prime.
- **Complexity:** $O(N \sqrt{N})$ because for every $N$ numbers we evaluate, we loop $\sqrt{N}$ times.

### ***Level up***
If you are generating cryptography keys or processing datasets to 1,000,000, this nested loop is too slow. DevOps engineers use an ancient Greek algorithm called the **Sieve of Eratosthenes**. Instead of checking if each room is empty, it walks through and preemptively crosses out all multiples (2s, 3s, 5s), leaving only the primes untouched.

````python
# The Sieve of Eratosthenes (Ultra-fast generation)
def generate_primes(limit):
    sieve = [True] * (limit + 1)
    for p in range(2, int(limit ** 0.5) + 1):
        if sieve[p]:
            # Cross out all multiples of p
            for i in range(p * p, limit + 1, p):
                sieve[i] = False
                
    return [p for p in range(2, limit + 1) if sieve[p]]

print("Fast Primes:", generate_primes(100))
````

### ***Edge-Case Test:***
Notice both algorithms start the range at `2`. The number `1` is neither prime nor composite. Including it in automated datasets will ruin prime factorization logic down the line.

**Task:** Enter a string and substring, then print how many times substring occurs from left to right.[cite: 5]

### ***The Plagiarism Checker***

### The Analogy
You are reviewing an essay you wrote at 3 AM. You want to make sure you didn't accidentally overuse the filler word "CDC". The script acts as a plagiarism scanner, reading the document left to right.
````python
main_string = "ABCDCDC"
sub_string = "CDC"

count = 0
start = 0

while True:
    # .find() returns the index of the first match, or -1 if not found
    idx = main_string.find(sub_string, start)
    if idx == -1:
        break
    count += 1
    #Shift the starting point forward by 1 to catch overlapping matches
    start = idx + 1

print("Occurence: ", count)
````

### Under the hood:

- **Gotchas:** The logic `start = idx + 1` is brilliant because it catches **overlapping substrings**. In `"ABCDCDC"`, the string `"CDC"` appears twice, overlapping at the middle 'C'. Python's native `main_string.count("CDC")` would return `1` because it skips over the overlap. This manual `.find()` loop catches both!
- **Complexity:** $O(N \times M)$ worst-case time complexity, but Python’s `.find()` uses the highly optimized Boyer-Moore-Horspool algorithm under the C hood.

### ***Level up***
If you are doing advanced string matching (like DNA sequence analysis or complex log parsing), you abandon `.find()` entirely and step up to **Regular Expressions (Regex)**. The `re` module is the industry standard for pattern hunting.

````python
import re

main_string1 = "ABCDCDC"
# Using regex with lookahead (?=...) to find overlapping matches
matches = re.findall(r'(?=CDC)', main_string1)

print(f"Regex Occurrences: {len(matches)}")
````

### ***Edge-Case Test:***

### ***Level up***
If you are analyzing millions of stock ticks for *Project Vegapunk*, sorting the entire dataset just to find the top two numbers is incredibly wasteful. You can find the runner-up in a single pass $O(N)$ by keeping track of the gold and silver medals dynamically.

````python
scores = [2, 3, 6, 6, 5]
gold = silver = float('-inf')

for score in scores:
    if score > gold:
        silver = gold      # Old gold becomes silver
        gold = score       # New gold is crowned
    elif gold > score > silver:
        silver = score     # New silver crowned, gold untouched

print(f"O(N) Fast Runner-Up: {silver}")
````

## Question 1: Max and Min without built-ins

**Task:** Write a function to find maximum and minimum from a sequence without using built-in `max()` and `min()`.[cite: 7]

### ***The Midterm Grader***

### The Analogy
You are the Teaching Assistant grading midterms for the intro coding class. You don't have a magical button to instantly know the highest and lowest scores. Instead, you pick up the very first paper and declare it both the "highest" and "lowest" score you've seen so far. Then, you read through the rest of the stack one by one. If a paper is higher than your current top score, it takes the crown. If it's lower than the bottom score, it takes that crown.

````python
def find_max_min(sequence):
    # Setting the baseline with the very first item
    maximum = sequence[0]
    minimum = sequence[0]

    # Looping through the rest of the items
    for value in sequence[1:]:
        if value > maximum:
            maximum = value
        if value < minimum:
            minimum = value

    return maximum, minimum

nums = [15, 3, 99, 42, 7]
mx, mn = find_max_min(nums)

print("Maximum:", mx)
print("Minimum:", mn)
````

### Under the hood:

- **Gotchas:** A huge beginner mistake is initializing `maximum = 0`. What if the sequence is `[-5, -10, -20]`? The maximum would stubbornly stay `0`, which isn't even in the list! Initializing with `sequence[0]` makes the logic bulletproof regardless of the data's range.
- **Complexity:** $O(N)$ time complexity. We traverse the sequence exactly once. $O(1)$ space complexity since we only store two tracking variables in memory.

### ***Level up***
In big data pipelines, sorting data is expensive. If you need the min and max to normalize data (like Min-Max Scaling in Machine Learning), doing it in one single pass $O(N)$ is optimal. We can improve our function's safety by handling cases where the data pipeline accidentally sends us an empty array.

````python
def safe_find_max_min(sequence):
    if not sequence:
        raise ValueError("Data pipeline Error: Sequence is empty.")
    
    # Unpacking via an iterator to be more memory efficient
    iterator = iter(sequence)
    maximum = minimum = next(iterator)
    
    for value in iterator:
        if value > maximum: maximum = value
        elif value < minimum: minimum = value # 'elif' saves CPU cycles
            
    return maximum, minimum

print("Safe Extracted Tuple:", safe_find_max_min([15, 3, 99, 42, 7]))
````

### ***Edge-Case Test:***
If you pass a list with identical numbers (e.g., `[5, 5, 5]`), the initial conditions set `maximum = 5` and `minimum = 5`. The loop conditions `>` and `<` will simply ignore the rest, correctly outputting `(5, 5)`.

## Question 2: Sum of cubes below n

**Task:** Write a function that takes a positive integer and returns sum of cubes of all positive integers smaller than it.[cite: 7]

### ***The Server Rack Capacity***

### The Analogy
You are the lead tech configuring the new AI server room. Rack 1 takes $1^3$ cooling units, Rack 2 takes $2^3$ cooling units, Rack 3 takes $3^3$, and so on. Before you install the massive $n$th rack, you need a function to dynamically calculate the cumulative cooling capacity already being consumed by all the smaller racks on the floor combined.

````python
def sum_of_cubes_below(n):
    total = 0
    # range(1, n) iterates from 1 up to (n-1), perfectly excluding 'n'
    for i in range(1, n):
        total += i ** 3
    return total

n = 6
print(f"Sum of cubes below {n}:", sum_of_cubes_below(n))
````

### Under the hood:

- **Gotchas:** The wording strictly says "smaller than it". By utilizing Python's `range(1, n)`, the function is natively exclusive at the top boundary. If you used `range(1, n+1)`, you would incorrectly sum the $n$th cube as well.
- **Complexity:** $O(N)$ runtime. It requires iterating through every number and doing an arithmetic operation.

### ***Level up***
Running a `for` loop to calculate mathematical sequences is a waste of CPU cycles. Using pure mathematics (Faulhaber's formula), we can calculate the sum of consecutive cubes using a formula that drops the runtime from $O(N)$ to an instant $O(1)$.
Formula: $S = (\frac{m(m+1)}{2})^2$ (where $m = n - 1$).

````python
def fast_sum_of_cubes(n):
    if n <= 1: return 0
    m = n - 1
    # Pure O(1) math, bypassing the loop completely
    return (m * (m + 1) // 2) ** 2

print(f"O(1) Math Sum:", fast_sum_of_cubes(6))
````

### ***Edge-Case Test:***
If `n = 1`, the problem asks for positive integers strictly smaller than 1. Since none exist, the loop `range(1, 1)` will instantly exit and correctly return `0`.

## Question 3: Recursive print 1 to n

**Task:** Write a recursive function to print numbers from `1` to `n` (no loops).[cite: 7]

### ***The Chain of Command***

### The Analogy
Imagine the chain of command for getting a new club approved on campus. 
1. The freshman ($n=3$) hands the form to the sophomore ($n=2$).
2. The sophomore hands it to the junior ($n=1$).
3. The junior reaches the faculty advisor (the Base Case, $n=0$).
The advisor approves it and hands it back. The junior stamps it (`print(1)`), the sophomore stamps it (`print(2)`), and the freshman finally gets it back and stamps it (`print(3)`). Recursion burrows all the way to the bottom *before* it starts doing the actual work on the way back up.

````python
def print_1_to_n(n):
    # The Base Case: The Faculty Advisor stopping the chain
    if n == 0:
        return
    
    # Burrowing down: Calling the function BEFORE printing
    print_1_to_n(n - 1)
    
    # Working on the way back up
    print(n, end=" ")

print_1_to_n(10)
print()
````

### Under the hood:

- **Gotchas:** The placement of `print(n)` dictates the direction! If you place `print(n)` *before* the recursive call `print_1_to_n(n-1)`, it prints on the way down, resulting in `10 9 8 7...` instead.
- **Complexity:** $O(N)$ time complexity. But more importantly, $O(N)$ **Space Complexity**. Every recursive call gets stacked in the computer's RAM (Call Stack).

### ***Level up***
Standard recursion crashes if $N$ is too large (like $N=2000$). Python has a default recursion depth limit of 1000 to prevent your RAM from overflowing. If you are building deep tree-search algorithms in AIML, you must manually override this limit using the `sys` module, but proceed with extreme caution.

````python
import sys

# Checking the default safety limit
print(f"Default Recursion Limit: {sys.getrecursionlimit()}")

# In extreme DevOps use-cases, we can raise it
sys.setrecursionlimit(2500)
print(f"New Limit Set: {sys.getrecursionlimit()}")
````

### ***Edge-Case Test:***
If `n` is negative (e.g., `-5`), `n == 0` will never trigger. The function will plunge downwards (`-6, -7, -8...`) until it triggers a fatal `RecursionError`. A robust base case should check `if n <= 0: return`.

## Question 4: Recursive Fibonacci

**Task:** Write a recursive function to print Fibonacci series up to `n` terms.[cite: 7]

### ***The Campus Rumor Pyramid***

### The Analogy
Think of how a rumor spreads during finals week. You want to confirm the full story. To do so, you have to ask your two closest friends. But they don't know the full story either, so they each have to ask their two closest friends. This pyramid expands exponentially until it hits the original two freshmen (Base Case 0 and 1) who started it.

````python
def fib(n):
    # Base cases: The original freshmen
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Recursive case: Asking the two friends below you
    return fib(n - 1) + fib(n - 2)

terms = 8
print("Fibonacci series:")
for i in range(terms):
    print(fib(i), end=" ")
print()
````

### Under the hood:

- **Gotchas:** This looks elegant, but mathematically, it is a disaster. To calculate `fib(5)`, it has to calculate `fib(3)` twice, `fib(2)` three times, etc. The tree duplicates its work massively.
- **Complexity:** $O(2^N)$ Exponential Time. If you try to calculate `fib(40)`, your computer will freeze for several seconds. `fib(50)` might take hours.

### ***Level up***
Data structures utilize a technique called **Memoization** to cure this exponential nightmare. We give the function a "memo pad" (a cache). Before it asks a friend for a number, it checks the memo pad. If it has already calculated `fib(10)` earlier, it instantly retrieves the answer instead of splitting the universe into two new recursive branches.

````python
from functools import lru_cache

# Caches the results of the recursive calls natively in memory
@lru_cache(maxsize=None)
def fast_fib(n):
    if n < 2: return n
    return fast_fib(n - 1) + fast_fib(n - 2)

# Instantly computes massive numbers, cutting O(2^N) down to O(N)
print(f"Fibonacci term 100: {fast_fib(100)}")
````

### ***Edge-Case Test:***
Asking for negative terms (`fib(-1)`) will result in maximum recursion depth exceeded, as it steps downwards infinitely (`-2, -3, -4`). A protective wrapper checking `if n < 0: raise ValueError` is mandatory for production.

## Question 1: Name metrics from a file

**Task:** Add a few names in `name.txt`, then count total names, count names starting with a vowel, and find the longest name.[cite: 8]

### ***The Dorm Party Guest List***

### The Analogy
You are hosting a massive dorm party and writing the guest list down on a physical clipboard (the `.txt` file). If you just kept the names in your head (RAM), you'd forget them the moment you went to sleep. By writing them to the hard drive, the data persists. Later, the bouncer (your script) reads the clipboard line-by-line to figure out how many people are coming, who gets VIP access (vowels), and whose name is the hardest to pronounce (the longest string).

````python
names = ["Aarav", "Ishita", "Rahul", "Omkar", "Esha", "Vivek"]

# 1. Writing to the clipboard (File)
with open("name.txt", "w", encoding="utf-8") as file:
    for name in names:
        file.write(name + "\n")

# 2. Bouncer reading the clipboard
with open("name.txt", "r", encoding="utf-8") as file:
    # .strip() removes the hidden \n newline characters from the file
    stored_names = [line.strip() for line in file if line.strip()]

vowels = tuple("AEIOUaeiou")
count_names = len(stored_names)
count_vowel_start = sum(1 for name in stored_names if name.startswith(vowels))
longest_name = max(stored_names, key=len) if stored_names else None

print("Names in file:", stored_names)
print("Count of names:", count_names)
print("Names starting with vowel:", count_vowel_start)
print("Longest name:", longest_name)
````

### Under the hood:

- **Gotchas:** When writing to a file, Python doesn't automatically add newlines. If you forget `+ "\n"`, your file will look like `AaravIshitaRahul`. When reading, that `\n` is still attached to the string! You must use `.strip()` to cleanly extract the name.[cite: 8]
- **Complexity:** $O(N)$ where $N$ is the number of lines in the file. `with open()` is a **Context Manager**—it guarantees the file is safely closed when the block ends, even if the program crashes halfway through.

### ***Level up***
If your party guest list has 10 million names, reading them all into a list `stored_names` at once will cause your computer's RAM to explode. To handle big data, DevOps engineers use **lazy evaluation**. We read and process the file exactly one line at a time, keeping RAM usage near zero.

````python
# Big Data Optimization: Processing without storing the whole file in RAM
total_guests = 0
longest = ""

with open("name.txt", "r", encoding="utf-8") as file:
    for line in file:
        clean_name = line.strip()
        if not clean_name: continue
        
        total_guests += 1
        if len(clean_name) > len(longest):
            longest = clean_name

print(f"Memory-Safe Total: {total_guests} | Longest: {longest}")
````

### ***Edge-Case Test:*** 
If the file is completely empty, Python's built-in `max()` function will trigger a fatal `ValueError: max() arg is an empty sequence`. The code utilizes an inline guard `max(...) if stored_names else None` to safely bypass this crash.

### The Analogy
You downloaded an old-school text database of university campuses. The data isn't in a nice Excel sheet; it's just raw rows separated by spaces. You have to write an extraction tool that chops each row into three distinct pieces (Name, Population, Area), converts the math parts into floats, and structures them properly so you can actually run analytical queries on them.

````python
city_records = [
    ("Dehradun", 5.78, 308.20),
    ("Delhi", 190.00, 1484.00),
    ("Mumbai", 124.00, 603.40),
]

# Simulating the raw text dump creation
with open("city.txt", "w", encoding="utf-8") as file:
    for city, population, area in city_records:
        file.write(f"{city} {population} {area}\n")

parsed_cities = []
# The Extraction Engine
with open("city.txt", "r", encoding="utf-8") as file:
    for line in file:
        parts = line.split() # Chops the line by spaces into a list
        if len(parts) == 3:
            city = parts[0]
            population = float(parts[1]) # Casting to math objects
            area = float(parts[2])
            parsed_cities.append((city, population, area))

print("All extracted details:")
for city, population, area in parsed_cities:
    print(f"{city}: pop={population} lakh, area={area} sq km")

# Querying the structured data
megacities = [city for city, population, _ in parsed_cities if population > 10]
total_area = sum(area for _, _, area in parsed_cities)

print("\nMegacities (>10 lakhs):", megacities)
print("Sum of all areas:", round(total_area, 2), "sq km")
````

The unpacking logic will shatter with a `ValueError: too many values to unpack`.
- **Complexity:** $O(N)$ string splitting and float conversion.

### ***Level up***
Manually splitting text files is brittle. If you are dealing with real-world tabular data (CSV, TSV), use Python's built-in `csv` module. It safely handles edge cases like spaces inside names if the strings are properly quoted (e.g., `"New York",8.4,783`).

````python
import csv

# Professional CSV parsing handles internal spaces and quotes natively
def parse_csv_safely(filepath):
    results = []
    with open(filepath, 'r') as csvfile:
        # Delimiter can be spaces or commas
        reader = csv.reader(csvfile, delimiter=' ')
        for row in reader:
            if len(row) == 3:
                results.append({"city": row[0], "pop": float(row[1])})
    return results

print("CSV Parser Active.")
````

### ***Edge-Case Test:*** 
The `_` in `for city, population, _ in parsed_cities` is a "throwaway" variable. It tells Python and other developers: "I am unpacking three variables to avoid structural errors, but I have absolutely no intention of using the 3rd variable (area) in this specific logic block."

````python
raw_input_data = "3\n1 0\n2 $\n3 1"

# Simulating reading lines from an input file or standard input
lines = raw_input_data.strip().splitlines()
n = int(lines[0])

for line in lines[1:n + 1]:
    a_text, b_text = line.split()
    
    # The Blast Shield
    try:
        result = int(a_text) // int(b_text)
        print("Success:", result)
    except ZeroDivisionError as error:
        print("Error Code: Div by Zero! ->", error)
    except ValueError as error:
        print("Error Code: Bad Character! ->", error)
````

### Under the hood:

## Question 2: Advanced Methods and Aggregation

**Task:** Add methods to calculate the percentage, return the pass/fail result, and write an external function to find the class average.[cite: 9]

### ***The Automated Grading Pipeline***

### The Analogy
Instead of having the professors manually calculate if a student passed their AIML classes, you bake the grading logic directly into the Student template. Now, every single ID card basically acts like a smart-contract that can grade itself. The `class_average` function acts as the Dean's dashboard, sweeping across all the individual smart-cards to generate a global metric.

````python
class StudentRecord:
    def __init__(self, name, sap_id, marks):
        self.name = name
        self.sap_id = sap_id
        self.marks = marks

    def display(self):
        print(f"[{self.sap_id}] {self.name} | Marks: {self.marks}")

    def marks_percentage(self):
        # Dynamically calculates based on the length of the dictionary
        return sum(self.marks.values()) / len(self.marks)

    def result(self):
        # all() evaluates True only if every single mark is > 40
        return "Pass" if all(mark > 40 for mark in self.marks.values()) else "Fail"

# The Aggregator Function (External to the class)
def class_average(records):
    total_percentage = sum(record.marks_percentage() for record in records)
    return total_percentage / len(records)

# Pipeline Data
records = [
    StudentRecord("Ray", "500069944", {"phy": 78, "chem": 67, "maths": 81}),
    StudentRecord("Kunal", "5002", {"phy": 88, "chem": 92, "maths": 85}),
    StudentRecord("Meera", "5003", {"phy": 39, "chem": 73, "maths": 64}) # Physics < 40
]

for record in records:
    print(f"{record.name} | Score: {record.marks_percentage():.2f}% | Status: {record.result()}")

print(f"\n[Global Class Average]: {class_average(records):.2f}%")
````

### Under the hood:

- **Gotchas:** Calling `len(self.marks)` assumes the dictionary isn't empty. If a student enrolled but has zero grades `{}` entered yet, `len()` returns 0, and the script violently crashes with a `ZeroDivisionError`.[cite: 9]
- **Complexity:** `marks_percentage()` operates in $O(M)$ where $M$ is the number of subjects. `class_average()` operates in $O(S \times M)$ where $S$ is the number of students.

### ***Level up***
In professional codebases, calling `student.marks_percentage()` as a function feels clunky because a percentage is an *attribute* of the student, not an *action* they are performing. We use the `@property` decorator to trick Python into treating the method exactly like a standard variable.

````python
class SmartStudent:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    @property
    def percentage(self):
        if not self.marks: return 0.0 # Guard clause against division by zero!
        return sum(self.marks.values()) / len(self.marks)

student = SmartStudent("Ray", {"AIML": 95, "DevOps": 88})
# Notice there are no parentheses () after percentage! It acts like a variable.
print(f"Dynamic Property Access: {student.percentage}%")
````

### ***Edge-Case Test:***
If you pass an empty dictionary to `all(mark > 40 for mark in self.marks.values())`, the `all()` function natively evaluates to `True`. Meaning a student with literally zero grades would default to "Pass". You must add `if not self.marks: return "Fail"`.

## Question 3: Types of Inheritance

**Task:** Create programs to implement single, multilevel, and multiple inheritance.[cite: 9]

### ***The Campus Persona Stack***

### The Analogy
Inheritance is about passing down abilities natively. 
*   **Single Inheritance:** The `DevOpsStudent` inherits basic terminal access from the overarching `UniversityStudent` class.
*   **Multilevel Inheritance:** Hardware hierarchy. A base `TechDevice` passes power functions down to a `MacSystem`, which passes macOS functions down to your specific `MacBookM5`.
*   **Multiple Inheritance:** You are the synergy of your disciplines. You inherit deep technical knowledge from your `AIMLMajor` class, and you inherit physical strength from your `FitnessAthlete` class, combining them both into your `ByteAndBarbell` persona.

````python
# --- 1. Single Inheritance ---
class UniversityStudent:
    def access_library(self):
        print("[Single] Accessing main campus library database.")

class DevOpsStudent(UniversityStudent):
    def deploy_server(self):
        print("[Single] Deploying AWS EC2 instance.")

# --- 2. Multilevel Inheritance ---
class TechDevice:
    def power_on(self):
        print("[Multilevel] Device booting up...")

class MacSystem(TechDevice):
    def load_macos(self):
        print("[Multilevel] Loading macOS kernel...")

class MacBookM5(MacSystem):
    def train_model(self):
        print("[Multilevel] Training AIML models on M5 silicon...")

# --- 3. Multiple Inheritance ---
class AIMLMajor:
    def code_pipeline(self):
        print("[Multiple] Architecting multi-model data pipelines.")

class FitnessAthlete:
    def lift_heavy(self):
        print("[Multiple] Executing 5x5 compound sets.")

class ByteAndBarbell(AIMLMajor, FitnessAthlete):
    def create_content(self):
        print("[Multiple] Recording authentic voiceovers for the audience.")

# Execution
print("--- Architecture Demos ---")
dev_ops = DevOpsStudent()
dev_ops.access_library() # Inherited

laptop = MacBookM5()
laptop.power_on() # Inherited from Grandparent
laptop.load_macos() # Inherited from Parent

ray = ByteAndBarbell()
ray.code_pipeline() # Inherited from Parent 1
ray.lift_heavy() # Inherited from Parent 2
ray.create_content() # Native method
````

### Under the hood:

- **Gotchas:** Multiple inheritance triggers the infamous **Diamond Problem**. If both `AIMLMajor` and `FitnessAthlete` possessed a method called `execute()`, which one would the child class use? Python relies on **MRO (Method Resolution Order)**, reading the parents strictly from left-to-right.[cite: 9]
- **Complexity:** Method lookups climb the inheritance tree dynamically, taking $O(D)$ time where $D$ is the depth of the inheritance tree.

### ***Level up***
If a child class overrides its parent's `__init__` constructor, the parent class's properties are destroyed unless explicitly salvaged. You must use `super().__init__()` to securely pass the initialization variables back up the chain.

````python
class BaseStudent:
    def __init__(self, name):
        self.name = name

class GradStudent(BaseStudent):
    def __init__(self, name, thesis_topic):
        # We MUST call super() so the Parent handles the 'name' logic
        super().__init__(name)
        self.thesis_topic = thesis_topic

student = GradStudent("Ray", "Automated Trading Algorithms")
print(f"Inherited Name: {student.name} | Thesis: {student.thesis_topic}")
````

### ***Edge-Case Test:***
You can instantly check a class's internal resolution logic by printing `ByteAndBarbell.mro()`. It outputs the exact sequence Python will follow when searching for a requested method, allowing you to debug inheritance conflicts flawlessly.

### ***Project Vegapunk's Upgrade***

When the system calls that method, your advanced AI implementation crushes and replaces the generic parent method.

````python
class BaseAlgorithm:
    def execute_trade(self):
        print("[Base Algo] Executing standard market order.")

class ProjectVegapunk(BaseAlgorithm):
    # This identically named method OVERRIDES the parent method
    def execute_trade(self):
        print("[Vegapunk Algo] Fetching morning data retrieval models...")
        print("[Vegapunk Algo] Running sentiment analysis. Executing paper trade.")

generic_bot = BaseAlgorithm()
advanced_bot = ProjectVegapunk()

print("--- Triggering Base System ---")
generic_bot.execute_trade()

print("\n--- Triggering Upgraded System ---")
advanced_bot.execute_trade()
````

### Under the hood:

- **Gotchas:** This relies heavily on exact spelling. If you accidentally name the child method `execute_trades()` (with an 's'), it won't override anything! It will just create a completely new method, and if you call `execute_trade()`, the parent's generic code will unexpectedly execute.[cite: 9]
- **Complexity:** Python checks the namespace of the current object first. Finding the method immediately resolves it in $O(1)$ without needing to traverse the MRO tree.

### ***Level up***
In large architectures, how do you force a teammate to override a method so they don't accidentally rely on the generic version? You use the `abc` (Abstract Base Class) module. It turns the parent into a strict contract that will crash the code if the child fails to supply an override.

````python
from abc import ABC, abstractmethod

class StrictBlueprint(ABC):
    @abstractmethod
    def mandatory_override(self):
        pass # This forces any child class to build this method

class CompliantChild(StrictBlueprint):
    def mandatory_override(self):
        print("Contract fulfilled. System stable.")

compliant = CompliantChild()
compliant.mandatory_override()
````

### ***Edge-Case Test:***

## Question 5: Operator Overloading

**Task:** Create a class for operator overloading which adds two `Point` objects where each point has `x` and `y` values.[cite: 9]

### ***The Campus Drone Pathing***

### The Analogy
Python natively understands that `5 + 5` equals `10`. But it has absolutely no idea what it means when you type `LocationA + LocationB`. You are programming a navigation drone that moves on an `(x, y)` grid across the campus quad. By "overloading" the `+` operator, you are essentially wiring the `+` button on your keyboard to a custom motor function that adds the GPS coordinates together logically.

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Magic Dunder Method: Hijacks the '+' symbol
    def __add__(self, other):
        # We create and return an entirely NEW Point object
        return Point(self.x + other.x, self.y + other.y)

    # Magic Dunder Method: Hijacks the print() function for this object
    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"

# P1 is the drone's start location, P2 is the movement vector
p1 = Point(10, 20)
p2 = Point(12, 15)

# Python sees '+', checks the object, and routes it to __add__
p3 = p1 + p2

print("Location 1:", p1)
print("Vector   2:", p2)
print("Final Loc :", p3)
````

### Under the hood:

- **Gotchas:** Magic methods (like `__add__`) should almost *always* return a brand new instance of the class rather than modifying `self.x` directly. If you modify `self.x`, calling `p3 = p1 + p2` will permanently corrupt `p1`'s data in the database.[cite: 9]
- **Complexity:** $O(1)$ computation.

### ***Level up***
What happens if someone accidentally types `p1 + 50`? The drone software will crash because `50` (an integer) doesn't have an `.x` or `.y` attribute. We level up our `__add__` method by injecting strict type verification.

````python
class SafePoint:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __add__(self, other):
        # Verifies the 'other' object is actually a SafePoint
        if not isinstance(other, SafePoint):
            raise TypeError("System Error: Can only add two SafePoint objects.")
        return SafePoint(self.x + other.x, self.y + other.y)

try:
    SafePoint(5, 5) + 100
except TypeError as error:
    print(error)
````

### ***Edge-Case Test:***
If you want to compare if two coordinates are exactly identical (e.g., `p1 == p2`), Python will return `False` even if the coordinates are both `(5,5)`. Why? Because they live in different memory locations. You must overload the equals operator `def __eq__(self, other):` to teach Python to compare their `.x` and `.y` values instead of their memory slots.
