### Under the hood
Unlike statically typed languages (where you have to strictly label the Tupperware `int weight = 45`), Python is dynamically typed. It looks at the item you put in the box and automatically slaps the correct label on it in memory.

### *Level Up -- Peeking inside the box*

The textbook question stops at "print the value and its type." I wanted to know: *how does Python actually store these differently in memory, and does `a = b` copy the box or just make a second label for the same box?* This distinction (identity vs equality, mutable vs immutable) trips up almost every beginner sooner or later, so it's worth seeing directly rather than being told about it.

```python
import sys

def inspect_value(label, value):
    print(f"{label:10s} | type ={type(value).__name__:6s} | id={id(value)} | size={sys.getsizeof(value)}B")

a = 10
b = 10           # small integers are cached in CPython, so a and b likely points to the same object
c = 10.0
d = [1, 2, 3]
e = d            # e is just another reference to the same list as d
f = list(d)      # f is a new list containing the same elements as d

inspect_value("a", a)
inspect_value("b", b)
print("a is b: ", a is b, "(CPython caches small ints -2 to 256 -- same box, two labels)")
inspect_value("c", c)
inspect_value("d", d)
inspect_value("e (=d)", e)
inspect_value("f (copy)", f)
print("d is e:", d is e, "  <- same object, mutating one mutates 'both'")
print("d is f:", d is f, "  <- different objects")
print("d == f:", d == f, " <- but equal VALUES")

d.append(99)
print("\nAfter d.append(99):")
print("d:", d, "| e:", e, " <- e changed too, because it's the SAME list")
print("f:", f, " <- f is untouched, it was a real copy")
```

**Why this matters in real code:** this exact gotcha (`e = d` vs `f = list(d)`) is one of the most common sources of bugs professional developers hit — passing a list into a function and being surprised the original changed outside it. Understanding it now, in Experiment 1, saves debugging pain in every experiment after this one.

### ***The Engineering Student's Calculator (Math & Geometry)***
Execute foundational arithmetic operations, evaluate algebraic expressions, and compute geometric formulas (Circle Area, Triangle Area, Hypotenuse).

### The Analogy
You and your study group are trying to calculate the absolute shortest path across the rectangular campus quad so you aren't late for `class (Hypotenuse)`. Then, you need to calculate the exact area of the pizzas you are ordering to ensure maximum food per dollar (Area formulas). Python's arithmetic operators are your digital calculator.

```python
import math

x, y = 9, 7
print(f"Basic Math: Add={x+y}, Sub={x-y}, Mult={x*y}, Div={x/y:.2f}") #

# Computing (x + y) * (x + y) with x=4, y=3
x_test, y_test = 4, 3 
expected_output = 49
actual_output = (x_test + y_test) * (x_test + y_test) #
print(f"Algebra: Expected {expected_output}, Got {actual_output}") #

# --- Q2: Area of a Pizza (Circle) ---
radius = 5.5 
circle_area = math.pi * (radius ** 2) 
print(f"Pizza Area: {circle_area:.2f} sq units")

# --- Q6: Shortest path across the Quad (Hypotenuse) ---
a, b = 3, 4
hypotenuse = math.sqrt(a**2 + b**2)
print(f"Quad Shortest Path (Hypotenuse): {hypotenuse}")

# --- Q8: Area of Campus Lawn (Triangle via Heron's Formula) ---
side_a, side_b, side_c = 3, 4, 5
s = (side_a + side_b + side_c) / 2 # Semi-perimeter
triangle_area = math.sqrt(s * (s - side_a) * (s - side_b) * (s - side_c)) #[cite: 2]
print(f"Lawn Area (Heron's): {triangle_area}")
```
