```python
#Largest number in an array with multiple indexes
arr = [1,2,4,6,4,6,4,4,5,6,4,6,8,8]

largest = arr[0]
indexes =[]
for i in range(1,len(arr)):
        if arr[i] > largest:
                largest = arr[i]

for i in range(len(arr)):
        if arr[i] == largest:
                indexes.append(i)

print(largest)
print(indexes)
```

```python
#Ascending order of the array
arr = [64, 34, 25, 12, 22, 11, 90]
for i in range(len(arr)):
    for j in range(0, i+1):
        if arr[i] < arr[j]:
            arr[i], arr[j] = arr[j], arr[i]
print(arr)
```

```python
# Rearrange names
arr = ['hello', 'world', 'python', 'programming']
for i in range(len(arr)):
    for j in range(0, i+1):
        if arr[i] < arr[j]:
            arr[i], arr[j] = arr[j], arr[i]
print(arr)
```

```python
# Rearrange names and then itself
arr = ['Aman', "Rohit", "Amit", "Rohit", "Aman", "Anshul", "binod"]

for i in range(len(arr)):
    char = list(arr[i])

    for j in range(len(char) -1):
        for k in range(len(char) -1):
            if ord(char[k]) > ord(char[k+1]):
                char[k], char[k+1] = char[k+1], char[k]
    arr[i] = ''.join(char)

print(arr)
```
